```python
def deleteStringColumns(df):
    col = list(df.select_dtypes(include='object').columns)
    df = df.drop(col,axis=1)
    return df
```

```python
d=deleteStringColumns(car)
df=d.drop(['car_ID','symboling'],axis=1)
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

# ModelBuilding

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X=df.drop(['price'],axis=1)
y=df['price']
```

```python
lr=LinearRegression(normalize=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=0)
y_test
```

```python
test=lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_test)
```

```python
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': pred})
df1
```

```python
print(lr.intercept_)
```

```python
df2 = pd.DataFrame(lr.coef_,X.columns,columns=['Coefficient'])
df2
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score=r2_score(y_test,pred)
r2_score
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

### OLS Method

```python
import statsmodels.api as sm
#splitting the dataset into train and test
X=sm.add_constant(X)
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=0)
print(X)
```

```python
import statsmodels.api as sm
from scipy import stats
model_ols=sm.OLS(y_train,X_train).fit()
```

```python
print(model_ols.summary())
```

```python
y_pred=model_ols.predict(X_train)
pd.DataFrame(y_pred,columns=['Train_data_predicted_values'])
```

```python
pd.DataFrame({'predicted_train_values':y_pred,'Actual__train_values':y_train})
```

```python
df_pred=model_ols.predict(X_test)
pd.DataFrame(df_pred,columns=['Test_data_predicted_values'])
```

```python
pd.DataFrame({'Predicted_test_values':df_pred,'Actual__test_values':y_test})
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, df_pred))
print('MSE:', metrics.mean_squared_error(y_test, df_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, df_pred)))
```

```python
print('MAE:', metrics.mean_absolute_error(y_train, y_pred))
print('MSE:', metrics.mean_squared_error(y_train, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, y_pred)))
```
